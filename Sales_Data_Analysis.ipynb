```python
#importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
#loading req dataset
df = pd.read_csv("/content/train_cleaned.csv")
```

```python
## Sales by category
sales_by_category = df.groupby("Category")["Sales"].sum().sort_values(ascending=False)
print("Sales by Category:\n", sales_by_category)
```

```python
# Sales by region
sales_by_region = df.groupby("Region")["Sales"].sum().sort_values(ascending=False)
print("\nSales by Region:\n", sales_by_region)
```

```python
# Monthly sales trend
df['Order Date'] = pd.to_datetime(df['Order Date'])
monthly_sales = df.set_index("Order Date").resample("ME")["Sales"].sum()
```

```python
# Export insights to CSV
sales_by_category.to_csv("sales_by_category.csv")
sales_by_region.to_csv("sales_by_region.csv")
monthly_sales.to_csv("monthly_sales.csv")
```

```python
# 1.visualizing  Sales by Category
plt.figure(figsize=(8,5))
sales_by_category.plot(kind="bar", color="skyblue")
plt.title("Sales by Category")
plt.ylabel("Total Sales")
plt.tight_layout()
plt.savefig("sales_by_category.png")
plt.show()
```

```python
# 2.visualizing Sales by Category
plt.figure(figsize=(6,6))
sales_by_region.plot(kind="pie", autopct="%1.1f%%")
plt.title("Sales by Region")
plt.ylabel("")
plt.tight_layout()
plt.savefig("sales_by_region.png")
plt.show()
```

```python
# 3.visualizing monthly sale trend
plt.figure(figsize=(10,5))
plt.plot(monthly_sales.index, monthly_sales.values, marker="o")
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter("%Y-%m"))
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.savefig("monthly_sales.png")
plt.show()
```
